```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Load the fruit image
img = cv2.imread('fruit.jpg')
plt.imshow(img)
plt.show()
```

```python
# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# Perform edge detection on the grayscale image
edges = cv2.Canny(gray, 100, 200)
```

```python
# Apply a Gaussian blur to the grayscale image
blurred = cv2.GaussianBlur(gray, (5, 5), 0)
```

```python
# Perform color-based segmentation on the original image
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
lower_range = np.array([0, 50, 50])
upper_range = np.array([10, 255, 255])
mask = cv2.inRange(hsv, lower_range, upper_range)
```

```python
# Apply morphological operations to the color-based mask
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
```

```python
# Combine the edge and color-based masks
mask = cv2.bitwise_or(mask, edges)
```

```python
# Find contours in the combined mask
contours, hierarchy = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
# Draw the contours on the original image
cv2.drawContours(img, contours, -1, (0, 255, 0), 2)
```

```python
# Display the result
plt.imshow(img)
plt.show()
```
